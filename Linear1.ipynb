```python
for col in df.columns:
    if df[col].dtype==object:
        df=df.fillna(df[col].mode())
    else:
        df=df.fillna(df[col].mean())
```

```python
for col in df.columns:
    if df[col].dtype!='object':
        sns.boxplot(df[col])
        plt.show()

```

```python
list=['bmi','past_consultations','num_of_steps','Hospital_expenditure','NUmber_of_past_hospitalizations','Anual_Salary']

for col in list:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)

    IQR=Q3-Q1

    LB=Q1-1.5*(IQR)
    UB=Q3+1.5*(IQR)

    df=df[(df[col]>=LB)&(df[col]<=UB)]
```

```python
for col in df.columns:
    if df[col].dtype!='object':
        sns.boxplot(df[col])
        plt.show()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for col in df.columns:
    if df[col].dtype =='object':
        df[col] = le.fit_transform(df[col].values)

print("Label encoding applied successfully.")
```
